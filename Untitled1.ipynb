```python
from sklearn.linear_model import LogisticRegression
x_train = [[30], [40], [50] ,[60], [70], [80]]
y_train = [0, 0, 1, 1, 1, 1]
model = LogisticRegression()
model.fit(x_train, y_train)
new_student_marks = [[44]]
prediction = model.predict(new_student_marks)
print("Prediction Output:", prediction)
if prediction[0] == 1:
  print("Student Result: PASS")
else:
  print("Student Result: FAIL")
```

```python
from sklearn.linear_model import LogisticRegression
x_train = [[30], [40], [50] ,[60], [70], [80]]
y_train = [0, 0, 1, 1, 1, 1]
model = LogisticRegression()
model.fit(x_train, y_train)
new_student_marks = [[44]]
prediction = model.predict(new_student_marks)
print("Prediction Output:", prediction)
if prediction[0] == 0:
  print("Student Result: PASS")
else:
  print("Student Result: FAIL")
```

```python
from sklearn.cluster import KMeans
X = [[35], [38], [50], [65], [68], [72], [85]]
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
print("Cluster Labels:", labels)
print("Cluster Centers:", kmeans.cluster_centers_)
for i in range(len(X)):
  print("Student with marks:", X[i][0],"-->","Belongs to Cluster:",labels[i])
```

```python
score=0
actions=["right", "right", "right", "left", "right"]
correct_action="right"
for step in range(len(actions)):
  action = actions[step]
  print("Step:", step + 1)
  if action == correct_action:
    score = score + 10
    print("Action:", action," ->Reward +10")
  else:
    score = score -5
    print("Action:", action,"->Penality-5")
  print("Current Score:", score)
  print("--------------")
print("Final Score of Agent:",score)
```
